```python
import pandas as pd
```

```python
data = pd.read_csv("hostel_choice.csv")
data.head()
```

```python
data.columns
```

```python
available_rooms = {}
numRoomsAddedList = pd.DataFrame(columns = ["batch", "gender", "hostel","numRoomsAdded"]);
for i in range(data.shape[0]):
    numRoomsAdded = 0

    row = data.iloc[i]
    batch = row["batch"].strip().lower()
    hostel = row['hostel'].strip().lower()
    gender = row['gender'].strip().lower()


    if(batch not in available_rooms.keys()):
        available_rooms[batch] = {}
    if(gender not in available_rooms[batch].keys()):
        available_rooms[batch][gender] = {};
    available_rooms[batch][gender][hostel] = {};

    floors = [int(floor.strip()) for floor in row["Floors"].split(",")]
    for floor in floors:
        available_rooms[batch][gender][hostel][floor] = []
    rooms = row['Rooms'];
    if rooms != "All":
        rooms = rooms.replace(")", "");
        rooms = rooms.replace(" ", "");
        roomsFloor = rooms.split("(")[1:]
        for roomF in roomsFloor:
            roomsRange = roomF.split("To");
            floor = int(roomsRange[0][0]);
            for room in range(int(roomsRange[0][1:]), int(roomsRange[1][1:]) + 1):
                if((hostel in ['kalam a', 'kalam d'] and str(room).endswith('8')) or (row["hostel"].strip().lower() == 'kalam b' and floor == 1 and room in list(range(1,10)))):
                    continue
                available_rooms[batch][gender][hostel][floor].append(room)
                numRoomsAdded += 1
    for floor in floors:
        # if floor is present but room data is not filled for that floor => all rooms in that floor
        if(not available_rooms[batch][gender][hostel][floor]):
            available_rooms[batch][gender][hostel][floor] = []
            for room in range(1, int(row['NumRoomsOnFloor'])+1):
                if((hostel in ['kalam a', 'kalam d'] and str(room).endswith('8')) or (row["hostel"].strip().lower() == 'kalam b' and floor == 1 and room in list(range(1,10)))):
                    continue
                available_rooms[batch][gender][hostel][floor].append(room)
                numRoomsAdded += 1


    print(batch,gender, hostel, ":", numRoomsAdded, "Rooms Added")
    numRoomsAddedList.loc[len(numRoomsAddedList)] = [batch,gender,hostel, numRoomsAdded]
    

```

```python
import json
with open('available_rooms.json', 'w') as fp:
    json.dump(available_rooms, fp)
```
